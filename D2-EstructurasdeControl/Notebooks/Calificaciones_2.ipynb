# Programa de calificaciones 
Este programa calcula el promedio de tres calificaciones y evalua si un alumno *aprobo* o no *aprobo* 

## Definiciono de variables 

```python
calif1 = 0.0
calif2 = 0.0
calif3 = 0.0
promedio= 0.0
```

```python
##Entrada de datos 
```

```python
calif1=float (input("Introduce la calificacion 1")) 
calif2=float (input("Introduce la calificacion 2")) 
calif3=float (input("Introduce la calificacion 3")) 
```

## PROCESO : CALCULO PROMEDIO

```python
promedio = (calif1 + calif2 + calif3)/3
```

## COMPARACION CON UNA ESTRUCTURA SELECTIVA DOUBLE 

```python
if promedio >= 7.0:
    input("Aprobado")
else:
    input ("No approbado")
```
